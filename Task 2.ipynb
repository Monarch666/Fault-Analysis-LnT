By Loadng the dataset and describing it we can see that the faults in G,C,B,A are in binary form so we can use that to determine the type of faults in the Electrical Grid

Thus first we have to convert this binary indicators

```python
df['Fault_Type'] = df[['G', 'C', 'B', 'A']].astype(str).agg(''.join, axis=1)

df
```

After defining it we can see how many types of faults are there in electrical grid through "Valuse count"

```python
fault_types = {
    '0000': 'No Fault',
    '1000': 'Single Line to Ground A',
    '0100': 'Single Line to Ground B',
    '0010': 'Single Line to Ground C',
    '0011': 'Line-to-Line BC',
    '0101': 'Line-to-Line AC',
    '1001': 'Line-to-Line AB',
    '1010': 'Line-to-Line with Ground AB',
    '0101': 'Line-to-Line with Ground AC',
    '0110': 'Line-to-Line with Ground BC',
    '0111': 'Three-Phase',
    '1111': 'Three-Phase with Ground',
    '1011': 'Line A Line B to Ground Fault'
}
```

```python
df['Fault_Type'] = df['Fault_Type'].map(fault_types)

df
```

```python
fault_type_counts = df['Fault_Type'].value_counts()
fault_type_counts
```

For visualisation and data distribution we will use Seaborn and matplotlib and use subplot to visualise

```python
features = ['Ia', 'Ib', 'Ic', 'Va', 'Vb', 'Vc']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

### Then we found the Correlation

```python
print("\nCorrelation Matrix:")
correlation_matrix = df[features].corr()
print(correlation_matrix)

# Here we used Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### Plotting the distribution of fault types

```python

plt.figure(figsize=(10, 10))
sns.countplot(y=df["Fault_Type"])
plt.title("Distribution of Fault Types")
plt.xlabel("Count")
plt.ylabel("Fault Type")
plt.show()
```

### Pairplot to observe pairwise relationships and distributions

```python
sns.pairplot(df[features])
plt.suptitle("Pairwise Relationships of Features", y=1.02)
plt.show()
```
